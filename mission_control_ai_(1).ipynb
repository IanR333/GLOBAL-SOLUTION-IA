## 🔧 PASSO 2 — Funções principais do sistema
Aqui definimos: geração de dados, lógica de alertas e conexão com a IA.

```python
import random
import ollama
from datetime import datetime

# ─────────────────────────────────────────────
# 1. GERAÇÃO DE DADOS SIMULADOS DA MISSÃO
# ─────────────────────────────────────────────
def gerar_dados_missao(cenario="normal"):
    """
    Gera dados simulados da missão espacial.
    cenario pode ser: 'normal', 'critico', ou 'aleatorio'
    """
    if cenario == "normal":
        temperatura  = random.uniform(18, 45)   # faixa segura: 0–60°C
        energia      = random.uniform(60, 100)  # faixa segura: > 30%
        comunicacao  = random.choice(["estável", "estável", "boa"])
        pressao      = random.uniform(95, 105)  # pressão em kPa
        modulo       = random.choice(["Módulo A", "Módulo B", "Módulo C"])

    elif cenario == "critico":
        temperatura  = random.uniform(85, 120)  # CRÍTICO: acima de 80°C
        energia      = random.uniform(5, 19)    # CRÍTICO: abaixo de 20%
        comunicacao  = random.choice(["instável", "falha", "sem sinal"])
        pressao      = random.uniform(60, 80)   # CRÍTICO: abaixo de 85 kPa
        modulo       = random.choice(["Módulo A", "Módulo B", "Módulo C"])

    else:  # aleatorio
        temperatura  = random.uniform(0, 120)
        energia      = random.uniform(0, 100)
        comunicacao  = random.choice(["estável", "boa", "instável", "falha", "sem sinal"])
        pressao      = random.uniform(60, 110)
        modulo       = random.choice(["Módulo A", "Módulo B", "Módulo C"])

    return {
        "modulo"       : modulo,
        "temperatura"  : round(temperatura, 1),
        "energia"      : round(energia, 1),
        "comunicacao"  : comunicacao,
        "pressao"      : round(pressao, 1),
        "timestamp"    : datetime.now().strftime("%d/%m/%Y %H:%M:%S")
    }


# ─────────────────────────────────────────────
# 2. LÓGICA DE ALERTAS (sem IA — só Python puro)
# ─────────────────────────────────────────────
def verificar_alertas(dados):
    """
    Verifica cada parâmetro e retorna lista de alertas.
    Essa é a lógica de tomada de decisão do sistema.
    """
    alertas = []

    # Regra 1: Temperatura
    if dados["temperatura"] > 80:
        alertas.append("🔴 ALERTA CRÍTICO: Temperatura acima de 80°C — risco de dano aos equipamentos!")
    elif dados["temperatura"] > 60:
        alertas.append("🟡 ATENÇÃO: Temperatura elevada — monitorar de perto.")

    # Regra 2: Energia
    if dados["energia"] < 20:
        alertas.append("🔴 ALERTA CRÍTICO: Energia abaixo de 20% — ativar modo de economia imediatamente!")
    elif dados["energia"] < 40:
        alertas.append("🟡 ATENÇÃO: Nível de energia baixo — verificar painéis solares.")

    # Regra 3: Comunicação
    if dados["comunicacao"] in ["sem sinal", "falha"]:
        alertas.append("🔴 ALERTA CRÍTICO: Comunicação perdida — iniciar protocolo de emergência!")
    elif dados["comunicacao"] == "instável":
        alertas.append("🟡 ATENÇÃO: Sinal instável — verificar antenas.")

    # Regra 4: Pressão
    if dados["pressao"] < 85:
        alertas.append("🔴 ALERTA CRÍTICO: Pressão crítica — possível vazamento detectado!")
    elif dados["pressao"] < 93:
        alertas.append("🟡 ATENÇÃO: Pressão abaixo do ideal.")

    if not alertas:
        alertas.append("✅ Todos os sistemas operando normalmente.")

    return alertas


# ─────────────────────────────────────────────
# 3. TOMADA DE DECISÃO AUTOMÁTICA
# ─────────────────────────────────────────────
def tomada_de_decisao(dados):
    """
    Com base nos dados, o sistema decide ações automáticas.
    """
    acoes = []

    if dados["energia"] < 20:
        acoes.append("⚡ AÇÃO: Modo de economia de energia ATIVADO automaticamente.")
        acoes.append("⚡ AÇÃO: Sistemas não essenciais DESLIGADOS.")

    if dados["temperatura"] > 80:
        acoes.append("❄️  AÇÃO: Sistema de resfriamento de emergência ATIVADO.")

    if dados["comunicacao"] in ["sem sinal", "falha"]:
        acoes.append("📡 AÇÃO: Protocolo de comunicação de backup INICIADO.")

    if dados["pressao"] < 85:
        acoes.append("🔒 AÇÃO: Compartimentos selados. Verificação de vazamento INICIADA.")

    if not acoes:
        acoes.append("🟢 Nenhuma ação automática necessária.")

    return acoes


# ─────────────────────────────────────────────
# 4. CONSULTA À IA (Llama via Ollama)
# ─────────────────────────────────────────────
def consultar_ia(dados, alertas):
    """
    Envia os dados e alertas para o Llama e pede uma análise.
    """
    # Monta o texto dos alertas para incluir no prompt
    alertas_texto = "\n".join(alertas)

    resposta = ollama.chat(
        model="llama3.2:1b",
        messages=[
            {
                "role": "system",
                # SYSTEM PROMPT: define o papel da IA dentro do contexto da missão
                "content": (
                    "Você é ARIA (Artificial Response Intelligence Assistant), "
                    "o sistema de IA de controle de missão espacial da nave FIAP-1. "
                    "Sua função é analisar os dados operacionais dos módulos da nave, "
                    "identificar riscos, e fornecer recomendações claras e objetivas "
                    "para a tripulação. Responda sempre em português, de forma técnica "
                    "mas compreensível. Máximo de 4 frases."
                )
            },
            {
                "role": "user",
                # USER PROMPT: os dados reais da leitura atual
                "content": (
                    f"Dados atuais do {dados['modulo']} — {dados['timestamp']}:\n"
                    f"• Temperatura: {dados['temperatura']}°C\n"
                    f"• Energia: {dados['energia']}%\n"
                    f"• Comunicação: {dados['comunicacao']}\n"
                    f"• Pressão: {dados['pressao']} kPa\n\n"
                    f"Alertas detectados pelo sistema:\n{alertas_texto}\n\n"
                    f"Forneça sua análise e recomendações para a missão."
                )
            }
        ]
    )

    return resposta["message"]["content"]


# ─────────────────────────────────────────────
# 5. EXIBIÇÃO DO PAINEL (output formatado)
# ─────────────────────────────────────────────
def exibir_painel(dados, alertas, acoes, analise_ia):
    """
    Exibe todos os dados de forma organizada no terminal/Colab.
    """
    linha = "═" * 55
    linha_fina = "─" * 55

    print(f"\n{linha}")
    print(f"   🚀 MISSION CONTROL AI — NAVE FIAP-1")
    print(f"   🛰️  {dados['modulo']} | {dados['timestamp']}")
    print(f"{linha}")

    print(f"\n  📊 LEITURAS DOS SENSORES")
    print(f"{linha_fina}")
    print(f"  🌡️  Temperatura  : {dados['temperatura']}°C")
    print(f"  ⚡  Energia      : {dados['energia']}%")
    print(f"  📡  Comunicação  : {dados['comunicacao']}")
    print(f"  🔵  Pressão      : {dados['pressao']} kPa")

    print(f"\n  ⚠️  ALERTAS DO SISTEMA")
    print(f"{linha_fina}")
    for alerta in alertas:
        print(f"  {alerta}")

    print(f"\n  🤖 AÇÕES AUTOMÁTICAS")
    print(f"{linha_fina}")
    for acao in acoes:
        print(f"  {acao}")

    print(f"\n  🧠 ANÁLISE DA IA (ARIA)")
    print(f"{linha_fina}")
    # Quebra o texto da IA em linhas de ~50 chars para ficar legível
    palavras = analise_ia.split()
    linha_atual = "  "
    for palavra in palavras:
        if len(linha_atual) + len(palavra) > 53:
            print(linha_atual)
            linha_atual = "  " + palavra + " "
        else:
            linha_atual += palavra + " "
    if linha_atual.strip():
        print(linha_atual)

    print(f"\n{linha}\n")


print("✅ Funções carregadas com sucesso!")
```

## ▶️ PASSO 3 — Rodar o sistema em cenário NORMAL

```python
print("Gerando leitura com dados normais...\n")

# 1. Gera os dados simulados
dados = gerar_dados_missao(cenario="normal")

# 2. Verifica alertas com base nos dados
alertas = verificar_alertas(dados)

# 3. Decide ações automáticas
acoes = tomada_de_decisao(dados)

# 4. Consulta a IA (Llama)
print("Consultando ARIA (IA)... aguarde...")
analise = consultar_ia(dados, alertas)

# 5. Exibe o painel completo
exibir_painel(dados, alertas, acoes, analise)
```

## 🚨 PASSO 4 — Rodar o sistema em cenário CRÍTICO
Agora simulamos uma emergência na missão!

```python
print("Simulando situação crítica na missão...\n")

dados_criticos = gerar_dados_missao(cenario="critico")
alertas_criticos = verificar_alertas(dados_criticos)
acoes_criticas = tomada_de_decisao(dados_criticos)

print("Consultando ARIA (IA)... aguarde...")
analise_critica = consultar_ia(dados_criticos, alertas_criticos)

exibir_painel(dados_criticos, alertas_criticos, acoes_criticas, analise_critica)
```

## 🔁 PASSO 5 — Monitoramento contínuo (3 leituras aleatórias)
Simula múltiplas leituras seguidas, como um sistema real faria.

```python
import time

print("🔄 Iniciando monitoramento contínuo — 3 ciclos de leitura...\n")

for ciclo in range(1, 4):
    print(f"{'─'*55}")
    print(f"  📡 CICLO DE LEITURA #{ciclo}")
    print(f"{'─'*55}")

    # Gera dados aleatórios (pode sair qualquer coisa)
    dados = gerar_dados_missao(cenario="aleatorio")
    alertas = verificar_alertas(dados)
    acoes = tomada_de_decisao(dados)

    print(f"Consultando ARIA... aguarde...")
    analise = consultar_ia(dados, alertas)

    exibir_painel(dados, alertas, acoes, analise)

    if ciclo < 3:
        print("⏳ Próxima leitura em 3 segundos...")
        time.sleep(3)

print("\n✅ Monitoramento concluído.")
```

```python
def chat_aria(pergunta):
    """Permite conversar livremente com a ARIA sobre a missão."""
    resposta = ollama.chat(
        model="llama3.2:1b",
        messages=[
            {
                "role": "system",
                "content": (
                    "Você é ARIA, a IA de controle da nave espacial FIAP-1. "
                    "Responda perguntas sobre a missão em português, "
                    "de forma técnica e objetiva. Máximo 3 frases."
                )
            },
            {"role": "user", "content": pergunta}
        ]
    )
    print("\n🧠 ARIA:", resposta["message"]["content"], "\n")

# Exemplos de uso — troque a pergunta pelo que quiser!
chat_aria("Quais são os principais riscos de uma missão de longa duração fora da Terra?")
```
